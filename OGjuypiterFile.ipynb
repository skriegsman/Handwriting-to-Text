```python
import cv2
import pytesseract
import matplotlib.pyplot as plt

image = cv2.imread(r"C:\Users\swkri\Desktop\pythonfile\Picture-to-Braille\pictures\IMG_1781.jpg", 0)

crop = image[1250:1785, 400:2700]
crop = cv2.GaussianBlur(crop, (9,9), 1.5)
ret, crop = cv2.threshold(crop,0,255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)

text = pytesseract.image_to_string(crop, config= '--psm 1 --oem 1')
print(text)

plt.figure(figsize=(20,10))
plt.imshow(cv2.cvtColor(crop, cv2.COLOR_BGR2RGB))
plt.show()

```
